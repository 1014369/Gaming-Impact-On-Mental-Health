```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("gaming_mental_health_10M_40features.csv")
df.head()
```

```python
df['Age_Group'] = 'Senior'
df.loc[df['age'] < 18, 'Age_Group'] = 'Teen'
df.loc[(df['age'] >= 18) & (df['age'] < 25), 'Age_Group'] = 'Young Adult'
df.loc[(df['age'] >= 25) & (df['age'] < 40), 'Age_Group'] = 'Adult'
df.loc[(df['age'] >= 40) & (df['age'] < 60), 'Age_Group'] = 'Middle Age'

print(df['Age_Group'].value_counts())
```

```python
df.columns
```

```python
df['Gaming_Intensity'] = 'Extreme'
df.loc[df['daily_gaming_hours'] <= 2, 'Gaming_Intensity'] = 'Casual'
df.loc[(df['daily_gaming_hours'] > 2) & (df['daily_gaming_hours'] <= 5), 'Gaming_Intensity'] = 'Regular'
df.loc[(df['daily_gaming_hours'] > 5) & (df['daily_gaming_hours'] <= 8), 'Gaming_Intensity'] = 'Heavy'

print(df['Gaming_Intensity'].value_counts())
```

```python

df['health_score'] = 0

df.loc[df['sleep_hours'] >= 6, 'health_score'] += 1
df.loc[df['exercise_hours'] >= 1, 'health_score'] += 1
df.loc[df['stress_level'] <= 5, 'health_score'] += 1
df.loc[df['anxiety_score'] <= 5, 'health_score'] += 1
df.loc[df['depression_score'] <= 5, 'health_score'] += 1
df.loc[df['addiction_level'] <= 5, 'health_score'] += 1
df.loc[df['bmi'] >= 18.5, 'health_score'] += 1
df.loc[df['bmi'] <= 29.9, 'health_score'] += 1
df.loc[df['happiness_score'] >= 5, 'health_score'] += 1
df.loc[df['loneliness_score'] <= 5, 'health_score'] += 1
df.loc[df['aggression_score'] <= 5, 'health_score'] += 1

df['Gamer_Health_Status'] = 'Unhealthy'
df.loc[df['health_score'] >= 5, 'Gamer_Health_Status'] = 'Moderate'
df.loc[df['health_score'] >= 9, 'Gamer_Health_Status'] = 'Healthy'

df.drop(columns=['health_score'], inplace=True)

print(df['Gamer_Health_Status'].value_counts())
```

```python
df['Sleep_Quality'] = 'Poor'
df.loc[df['sleep_hours'] >= 6, 'Sleep_Quality'] = 'Average'
df.loc[df['sleep_hours'] >= 8, 'Sleep_Quality'] = 'Good'

print(df['Sleep_Quality'].value_counts())
```

```python
df['BMI_Category'] = 'Obese'
df.loc[df['bmi'] < 18.5, 'BMI_Category'] = 'Underweight'
df.loc[(df['bmi'] >= 18.5) & (df['bmi'] < 25), 'BMI_Category'] = 'Normal'
df.loc[(df['bmi'] >= 25) & (df['bmi'] < 30), 'BMI_Category'] = 'Overweight'

print(df['BMI_Category'].value_counts())
```

```python
df['Income_Group'] = 'High'
df.loc[df['income'] < 20000, 'Income_Group'] = 'Low'
df.loc[(df['income'] >= 20000) & (df['income'] < 60000), 'Income_Group'] = 'Middle'
df.loc[(df['income'] >= 60000) & (df['income'] < 100000), 'Income_Group'] = 'Upper Middle'

print(df['Income_Group'].value_counts())
```

```python
df['Violent_Games_Level'] = 'High'
df.loc[df['violent_games_ratio'] < 0.33, 'Violent_Games_Level'] = 'Low'
df.loc[(df['violent_games_ratio'] >= 0.33) & (df['violent_games_ratio'] < 0.66), 'Violent_Games_Level'] = 'Medium'

print(df['Violent_Games_Level'].value_counts())
```

```python
df.to_csv('Gaming_Impact_On_Mental_Health.csv', index=False)
print("Exported successfully!")
print("Total Rows:", len(df))
print("Total Columns:", len(df.columns))
print("\nAll Columns:\n", df.columns.tolist())
```
